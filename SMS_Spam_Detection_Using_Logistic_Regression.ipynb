```python
# libraries
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score


# Step 1: Training data (REAL SMS-LIKE DATA)
messages = [
    "Win a free iPhone now",
    "Congratulations you won prize",
    "Call now to claim reward",
    "Hi how are you",
    "Let's meet tomorrow",
    "Are you coming today"
]

labels = [1, 1, 1, 0, 0, 0]
# 1 = Spam
# 0 = Not Spam


# Step 2: Convert text into numbers
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(messages)


# Step 3: Create classification model
model = LogisticRegression()


# Step 4: Train the model
model.fit(X, labels)


# Step 5: Test on training data
predictions = model.predict(X)


# Step 6: Check accuracy
accuracy = accuracy_score(labels, predictions)
print("Model Accuracy:", accuracy)


# Step 7: Predict for a NEW message
new_message = ["Win free money now"]
new_message_vector = vectorizer.transform(new_message)
prediction = model.predict(new_message_vector)

if prediction[0] == 1:
    print("This message is SPAM")
else:
    print("This message is NOT SPAM")
```
